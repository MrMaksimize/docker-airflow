```python
import pandas as pd
import requests
import json
from datetime import datetime
import math
```

```python
payload = {'format':'json'}
countRequest = requests.post('https://netfile.com:443/Connect2/api/public/list/filing', params=payload,
                    data = {'Aid':'CSD',
                            'Application': '',
                            'Form': '23',
                            'CurrentPageIndex':'0',
                            'PageSize':'1'
                           })
formTransactions = countRequest.json()['totalMatchingCount']

if formTransactions < 1000:
    requestLoops = 1
else:
    requestLoops = math.ceil(formTransactions/1000)
      
transactionsList = []

for i in range(requestLoops):
    page = str(i)
    transactionsRequest = requests.post('https://netfile.com:443/Connect2/api/public/list/filing', params=payload,
                    data = {'Aid':'CSD',
                            'Application': '',
                            'Form': '23',
                            'CurrentPageIndex':page,
                            'PageSize':'1000'
                           })
    allTransactions = transactionsRequest.json()['filings']
      
    for t in allTransactions:
        transactionsList.append([   t['id'], #schedule
                                       t['agency'], #recipient_id
                                       t['isEfiled'], #recipient_name
                                       t['hasImage'], #'report_period_from'
                                       t['filingDate'], #'report_period_to'
                                       t['title'], #'contributor_code'
                                       t['form'], #'contributor_last'
                                       t['filerName'] #'contributor_first'
                                       ])


campaignTransactions = pd.DataFrame(transactionsList,columns=['id','agency','isEfiled','hasImage','filingDate','title','form','filerName']) 
```
